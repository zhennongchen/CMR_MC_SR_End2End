```python
import os
import numpy as np
import nibabel as nb
import pandas as pd
import matplotlib.pyplot as plt
import CMR_HFpEF_Analysis.functions_collection as ff

main_path = '/mnt/mount_zc_NAS/HFpEF/'
save_path = os.path.join(main_path, 'pictures')
```

Low resolution

```python
data = pd.read_excel(os.path.join(main_path, 'data', 'Bspline_7-control-points_baseline_error_simulated_data_version2_LR.xlsx'))
```

```python
slice_num = [8,9,10,11,12]

error_x = np.mean(data.loc[data['num_slices'] == 12]['error_x'])
error_y = np.mean(data.loc[data['num_slices'] == 12]['error_y'])
error = np.mean(data.loc[data['num_slices'] == 12]['error'])
print(error_x, error_y, error)
```

```python
barWidth = 0.25
 
# set height of bar
error_x = [0.035, 0.043, 0.050, 0.122, 0.131]
error_y = [0.035, 0.054, 0.068, 0.135, 0.140]
error = [0.055, 0.078, 0.093, 0.201, 0.220]
 
# Set position of bar on X axis
br1 = np.arange(len(error_x))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make the plot
fig = plt.subplots(figsize =(6,4.5))
plt.bar(br1, error_x, color ='r', width = barWidth,
        edgecolor ='grey', label ='error_x')
plt.bar(br2, error_y, color ='g', width = barWidth,
        edgecolor ='grey', label ='error_y')
plt.bar(br3, error, color ='b', width = barWidth,
        edgecolor ='grey', label ='error_Euclidean')
 
# Adding Xticks
# plt.xlabel('Number of SAX slices',  fontsize = 14) #fontweight ='bold'
# plt.ylabel('Error (pixel)', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(error_x))],
        ['8', '9', '10', '11', '12'], fontsize = 15)

plt.yticks(np.arange(0.00,0.30,step = 0.05),fontsize=15)
# plt.title('Error by 7-CP B-spline interpolation in Low-resolution data')
plt.grid(linestyle = 'dashed')
 
# plt.legend()
plt.savefig(os.path.join(save_path, 'Error by 7-CP B-spline interpolation in Low-resolution data.eps'), format='eps')
```

```python
data = pd.read_excel(os.path.join(main_path, 'data', 'Bspline_10-control-points_baseline_error_simulated_data_version2_HR.xlsx'))
```
