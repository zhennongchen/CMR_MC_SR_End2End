```python
import numpy as np
import pydicom
import os
import h5py
import glob
import pandas as pd
import shutil

def find_all_target_files(target_file_name,main_folder):
    F = np.array([])
    for i in target_file_name:
        f = np.array(sorted(glob.glob(os.path.join(main_folder, os.path.normpath(i)))))
        F = np.concatenate((F,f))
    return F

def XX_to_ID_00XX(num):
    if num < 10:
        return 'ID_000' + str(num)
    elif num>= 10 and num< 100:
        return 'ID_00' + str(num)
    elif num>= 100 and num < 1000:
        return 'ID_0' + str(num)
    elif num >= 1000:
        return 'ID_' + str(num)
```

```python
# load ID_list
patient_list = pd.read_excel(os.path.join('/Users/zhennongchen/Documents/Zhennong_CMR_HFpEF/Data/HFpEF/Patient_list','Important_HFpEF_Patient_list_unique_patient_w_ZCnotes.xlsx'))
raw_data_list = pd.read_excel(os.path.join('/Users/zhennongchen/Documents/Zhennong_CMR_HFpEF/Data/HFpEF/Patient_list','raw_data_correspondence_HFpEF.xlsx'))
patient_list = patient_list.loc[patient_list['include?'] == 'yes']
print(patient_list.shape)
# define save folder
save_folder = '/Volumes/TOSHIBA_4TB/MGH/HFpEF_zhennong/dicoms_LAX'
```

```python
for i in range(0,patient_list.shape[0]):
    patient_num = patient_list.iloc[i]['OurID']
    print(patient_num)
    patient_id = XX_to_ID_00XX(patient_num)
    print(patient_id)
    
    raw_data_row = raw_data_list.loc[raw_data_list['OurID'] == patient_num].iloc[0]
    folder_path = str(raw_data_row['folder_path'])

    
    sub_folders = find_all_target_files(['*'],folder_path)
    for s in sub_folders:
        if '.xml' not in s:
            dcm_files = find_all_target_files(['*dcm'],s)
            if len(dcm_files) <= 10:
                continue
            dcm_meta = pydicom.dcmread(dcm_files[0])
            series = dcm_meta.SeriesDescription
            
            if 'CHAM' in series:
                # these are LAX folders we want
                save_folder_patient = os.path.join(save_folder, patient_id)
                os.makedirs(save_folder_patient, exist_ok = True)
                random_id = int(np.random.rand() * 8000) + 1000
               
                shutil.copytree(str(s), os.path.join(save_folder_patient, str(series) + '_' + str(random_id)))
    
    
```
